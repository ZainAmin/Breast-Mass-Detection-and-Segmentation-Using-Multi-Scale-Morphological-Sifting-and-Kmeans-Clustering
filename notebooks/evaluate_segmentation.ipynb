```python
import cv2
import numpy as np



def calculate_iou(contour1, contour2):
    mask1 = np.zeros_like(ground_truth_img, dtype=np.uint8)
    mask2 = np.zeros_like(segmented_img, dtype=np.uint8)

    if len(contour1) > 0:
        cv2.drawContours(mask1, [contour1], 0, 1, thickness=cv2.FILLED)

    if len(contour2) > 0:
        cv2.drawContours(mask2, [contour2], 0, 1, thickness=cv2.FILLED)

    intersection = np.logical_and(mask1, mask2)
    union = np.logical_or(mask1, mask2)
    iou = np.sum(intersection) / np.sum(union)
    return iou

def calculate_true_positive_rate(ground_truth_img, segmented_img, threshold=0.5):
    contours, _ = cv2.findContours(ground_truth_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    ground_truth_contours = [np.squeeze(contour) for contour in contours]

    contours, _ = cv2.findContours(segmented_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    segmented_contours = [np.squeeze(contour) for contour in contours]

    true_positives = 0
    total_ground_truth = len(ground_truth_contours)
    total_segmented = len(segmented_contours)

    for segmented_contour in segmented_contours:
        max_iou = 0

        for ground_truth_contour in ground_truth_contours:
            iou = calculate_iou(segmented_contour, ground_truth_contour)
            if iou > max_iou:
                max_iou = iou

        if max_iou >= threshold:
            true_positives += 1

    true_positive_rate = true_positives / total_ground_truth
    return true_positive_rate



# Example usage
ground_truth_img = cv2.imread(r"../dataset/output/upsampled/groundtruth/20587664_f4b2d377f43ba0bd_MG_R_ML_ANON.tif", cv2.IMREAD_GRAYSCALE)
segmented_img = cv2.imread(r"../dataset/output/upsampled/segmentation/20587664_f4b2d377f43ba0bd_MG_R_ML_ANON_1.tif", cv2.IMREAD_GRAYSCALE)


_, ground_truth_img = cv2.threshold(ground_truth_img, 1, 255, cv2.THRESH_BINARY)
_, segmented_img = cv2.threshold(segmented_img, 1, 255, cv2.THRESH_BINARY)


# print(ground_truth_img.shape, ground_truth_img.dtype)

tpr = calculate_true_positive_rate(ground_truth_img, segmented_img)
print("Region-wise global sensitivity:", tpr)
```

```python
def Metrics(cropped_img_upsampled, vector_gt_mask, contours_vector, contours_gt):
    IOU_vector = []

    for k in range(len(contours_gt)):
        max_iou = 0

        for contours in contours_vector:
            if len(contours) > 0:
                contour = contours[k]

                if len(contour) > 0:
                    mask_candidate = np.zeros_like(cropped_img_upsampled)
                    cv2.drawContours(mask_candidate, [contour], 0, (255, 255, 255), cv2.FILLED)

                    intersection = cv2.bitwise_and(mask_candidate, vector_gt_mask[k])
                    union = cv2.bitwise_or(mask_candidate, vector_gt_mask[k])

                    intersection_pixels = cv2.countNonZero(intersection)
                    union_pixels = cv2.countNonZero(union)

                    iou = intersection_pixels / union_pixels

                    if iou > max_iou:
                        max_iou = iou

        IOU_vector.append(max_iou)

    return IOU_vector
```
