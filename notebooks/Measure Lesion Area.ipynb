```python
import cv2
import os
import matplotlib.pyplot as plt

def measure_lesion_area(image_path):
    # Load the binary image
    image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

    threshold = 128
    _, image = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)

    # Find contours in the image
    contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Calculate the area of each contour
    areas = [cv2.contourArea(contour) for contour in contours]

    # Create a copy of the original image
    image_copy = image.copy()

    # Convert the image to RGB for plotting
    image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

    # Plot the original image with all contours
    plt.imshow(image_copy)
    plt.axis('off')
    plt.show()

    return areas

# Example usage
image_path = os.path.join(
    os.path.dirname("../dataset/processed/groundtruth/") ,
    '20587612_f4b2d377f43ba0bd_MG_R_CC_ANON.tif'
)


lesion_area = measure_lesion_area(image_path)
print(f"Lesion area: {lesion_area}")
```

```python
# Set the paths to the positive images and groundtruth folders
groundtruth_path = r'../dataset/processed/groundtruth'

# Get a list of all the tif images in the positive images folder
groundtruth_dirs = [os.path.join(
    os.path.dirname(r"../dataset/processed/groundtruth/") ,f) for f in os.listdir(groundtruth_path) if f.endswith('.tif')]

groundtruth_dirs
```

```python
area_list = []

for path in groundtruth_dirs:
    lesion_area = measure_lesion_area(path)
    area_list.append(lesion_area)
    print(f"path: {path}, Lesion area: {lesion_area} square pixels")
```

```python
import itertools

flattened_array = list(itertools.chain.from_iterable(area_list))
# print(max(flattened_array), min(flattened_array))

sorted_array = sorted(flattened_array)
print(sorted_array)

```

```python
# Calculate the average of all areas
average = sum(area_list) / len(area_list)

# Calculate the minimum and maximum average values
minimum_average = min(area_list, key=lambda x: abs(x - average))
maximum_average = max(area_list, key=lambda x: abs(x - average))

print(f"Minimum average value: {minimum_average}")
print(f"Maximum average value: {maximum_average}")
```
