```python
import os
import shutil

# Define the names of the two classes
classes = ['positive_images', 'negative_images']

# Create a folder for each class if it doesn't already exist
for class_name in classes:
    folder_path = os.path.join('../dataset', class_name)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
```

```python
# Read the file containing the names of the positive images
with open(os.path.join('../dataset/positives.txt'), 'r') as f:
    positive_image_names = f.read().splitlines()

# Move the positive images to the positive_images folder
for img_name in positive_image_names:
    img_path = os.path.join('../dataset/images/', img_name)
    dest_path = os.path.join('../dataset/positive_images', img_name)

    # shutil.copy(img_path, dest_path)
```

```python
# Read the file containing the names of the negative images
with open(os.path.join('../dataset/negatives.txt'), 'r') as f:
    negative_image_names = f.read().splitlines()

# Move the negative images to the negative_images folder
for img_name in negative_image_names:
    img_path = os.path.join('../dataset/images/', img_name)
    dest_path = os.path.join('../dataset/negative_images', img_name)

    # print(img_path)
    shutil.copy(img_path, dest_path)
```

```python
# Remove the empty images folder from the positive_images folder
images_folder_path = os.path.join('positive_images', 'images')
if os.path.exists(images_folder_path) and os.path.isdir(images_folder_path) and not os.listdir(images_folder_path):
    os.rmdir(images_folder_path)
```

```python
# Print out the names of all the positive images in the positive_images folder
positive_images_folder = os.path.join(os.getcwd(), 'positive_images')

# Get a list of all files in the positive_images folder
positive_image_files = os.listdir(positive_images_folder)

# Filter the list based on the file extension
positive_tif_files = [file for file in positive_image_files if file.endswith('.tif')]

# Print out the names of the positive images
for file in positive_tif_files:
    print(file)
```

```python
# Print out the names of all the positive images in the positive_images folder
negative_images_folder = os.path.join(os.getcwd(), 'negative_images')

# Get a list of all files in the negative_images folder
negative_image_files = os.listdir(negative_images_folder)

# Filter the list based on the file extension
negative_tif_files = [file for file in negative_image_files if file.endswith('.tif')]

# Print out the names of the negative images
for file in negative_tif_files:
    print(file)
```
